# K-Means Clustering: 

```python
from sklearn import datasets
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

%matplotlib inline
```

```python
iris = datasets.load_iris()
X = iris.data
df = pd.DataFrame(X)
df.columns = iris.feature_names
```

```python
df.head()
```

```python
def find_k (X, k_range, sample_percent=1):
    """
    k_range: a list of possible k values
    X, the data we're clustering on
    """
    from sklearn.cluster import KMeans
    import matplotlib.pyplot as plt
    import numpy as np
    from scipy.spatial.distance import cdist
    from sklearn.metrics import pairwise_distances

    N = X.shape[0]
    sampleSize = X.shape[0] * sample_percent

    if sampleSize > 0:
        index = np.arange(np.shape(X)[0])
        np.random.shuffle(index)
        X =  X[index, :]


    mean_distortions=[]
    for k in k_range:
        #cluster using k, then calculate the mean distortion (average distance to closest centroid)
        kmeans_model = KMeans(n_clusters=k, init='k-means++', n_jobs=-1).fit(X)
        mean_distortions.append(sum(np.min(pairwise_distances(X, kmeans_model.cluster_centers_,
                                                              metric='euclidean'),axis=1)) / X.shape[0])


    #visualize results
    plt.plot(k_range, mean_distortions)
    plt.xlabel("K Value")
    plt.ylabel("Mean Distortion")
    plt.title("Elbow Graph for Mean Distortion per K")
    plt.show()
```

```python
find_k(np.matrix(X), range(1,10),1)
```

From the above graph we can visualize that, there is a hitch at points when value of clusters are 2 and 3, so 2 and 3 would be the good option for value of K. So lets try with both values of K. 
(Note: Since I had already worked on this dataset before, I know there are 3 class values for Iris dataset, hence I guess 3 would be better option and will give us with better grouping.)

### When number of Cluster, K=2

```python
kmeans_model = KMeans(n_clusters=2, random_state=42)
kmeans_model.fit(df)
```

```python
df['labels'] = kmeans_model.labels_
```

```python
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

model = TSNE(n_components=3, learning_rate=12, random_state=10)
X_tsne = model.fit_transform(X) 
plt.figure(figsize=(10,10))
ax = plt.subplot(111, projection='3d')
ax.scatter(X_tsne[:, 0], X_tsne[:, 1],X_tsne[:,2], c=kmeans_model.labels_)
```

From the above grah we can see that the each data points belong to two clusters i.e. blue and red. Blue points that belong to one cluster are mostly scattered in lower part whereas red points that belong to the other cluster are scattered in upper part of the 3D plane.
